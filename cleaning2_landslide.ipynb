```python
l = pd.read_csv('./landslide_mod.csv')
GL = pd.read_csv('./GlobalLandslide_mod.csv')
```

```python
for i in range(len(l)):
    
    if (l['trigger'][i] == "Downpour"):
        l['trigger'][i] = 1
        
    if (l['trigger'][i] == "Rain"):
        l['trigger'][i] = 1    
        
    if (l['trigger'][i] == "Earthquake"):
        l['trigger'][i] = 2
        
    if (l['trigger'][i] == "Continuous_Rain"):
        l['trigger'][i] = 1   
        
    if (l['trigger'][i] == "Unknown"):
        l['trigger'][i] = 0 
    
```

```python
for i in range(len(l)):
    
    if (l['landslide_size'][i] == "Very_large"):
        l['landslide_size'][i] = 1
        
    if (l['landslide_size'][i] == "Large"):
        l['landslide_size'][i] = 2   
        
    if (l['landslide_size'][i] == "Medium"):
        l['landslide_size'][i] = 3
        
    if (l['landslide_size'][i] == "Small"):
        l['landslide_size'][i] = 4
```

```python
for i in range(len(l)):
    
    if (l['landslide_type'][i] == "Landslide"):
        l['landslide_type'][i] = 1
        
    if (l['landslide_type'][i] == "Mudslide"):
        l['landslide_type'][i] = 2   
        
    if (l['landslide_type'][i] == "Rockfall"):
        l['landslide_type'][i] = 3
```
